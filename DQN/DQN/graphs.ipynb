```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('acrobot')
episodes = [i for i in range(1, 1_000 + 1)]
plt.plot(
    episodes,
    data['train']
)
plt.title('Acrobot Training-Score vs Episode')
plt.xlabel('episode')
plt.ylabel('score')
plt.ylim(-2000, 0)
```
